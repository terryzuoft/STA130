````python
# Simulation Code 

import numpy as np

# Set the parameters
n_couples = 124  # Total number of couples
observed_tilts_right = 80  # Number of couples tilting to the right in observed data
n_simulations = 10000  # Number of simulations
p_right = 0.5  # 50/50 chance of tilting right or left under H0

# Simulate the experiment n_simulations times
simulated_tilts = np.random.binomial(n_couples, p_right, n_simulations)

# Calculate the p-value: the proportion of simulations where 80 or more couples tilted right
p_value = np.mean(simulated_tilts >= observed_tilts_right)

# Output the p-value
print(f"P-value: {p_value}")
````

**Null hypothesis:** couples have no preference for tilting their head left or right when kissing. 

**Alternative hypothesis:** couples have a preference for tilting their head left or right when kissing. 

After simulating with a few lines of Python code, we derived a p-value of 0.0004, using the table provided, we have very strong evidence against the null hypothesis. Where we can conclude that **it is very unlikely that couples have no preference for tilting their head left or right**, or in other words, it is very likely that couples have a preference for tilting their head left or right.

````python
# Modified Code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Reproducibility
np.random.seed(42)

# Reading the CSV data directly from GitHub raw URL
url = 'https://raw.githubusercontent.com/terryzuoft/STA130/main/vaccine_data.csv'
data = pd.read_csv(url)

# Calculate the difference in health scores
data['ScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Data Visualization
plt.figure(figsize=(10, 5))
sns.boxplot(data=data[['InitialHealthScore', 'FinalHealthScore']])
plt.title('Initial vs Final Health Scores')
plt.show()

# Bootstrapping function to resample data and calculate mean differences
def bootstrap(data, n_bootstrap=1000):
    bootstrap_means = []
    for i in range(n_bootstrap):
        sample = data.sample(frac=1, replace=True)
        mean_diff = sample['FinalHealthScore'].mean() - sample['InitialHealthScore'].mean()
        bootstrap_means.append(mean_diff)
    return bootstrap_means

# Run bootstrapping
bootstrap_means = bootstrap(data)

# Plot bootstrapping results
plt.hist(bootstrap_means, bins=30, edgecolor='black')
plt.axvline(np.mean(bootstrap_means), color='red', linestyle='dashed', linewidth=2)
plt.title('Bootstrapping Mean Differences')
plt.show()

# Findings: Calculate p-value for one-sided test
observed_mean_diff = data['ScoreDifference'].mean()

# CHANGE: Here we adjust for a one-sided hypothesis test (FinalHealthScore > InitialHealthScore)
p_value = np.mean(np.array(bootstrap_means) >= observed_mean_diff)

# Alternatively, for testing FinalHealthScore < InitialHealthScore, use:
# p_value = np.mean(np.array(bootstrap_means) <= observed_mean_diff)

print(f"Observed Mean Difference: {observed_mean_diff}")
print(f"p-value (one-sided test): {p_value}")
````



### What changed in the code 

1. **Bootstrapping Process**: The bootstrapping procedure remains the same. It resamples the data many times to generate a distribution of mean differences between `FinalHealthScore` and `InitialHealthScore`.

2. **p-value Calculation for One-Sided Test**:
   - I changed the p-value calculation to only consider one side of the distribution, based on whether we are testing if the final score is **greater than** the initial score.
   - The original code:
     ```python
     p_value = np.mean(np.array(bootstrap_means) >= observed_mean_diff)
     ```
     tests the hypothesis that the **final score is greater than the initial score**.
   - If we were testing the opposite hypothesis (final score **less than** initial score), we would modify it to:
     ```python
     p_value = np.mean(np.array(bootstrap_means) <= observed_mean_diff)
     ```
   - These are **one-sided tests**, contrasting with the previous two-sided test, where we would check if the difference was **either greater or less** than the observed difference, covering both tails.

### How this changes the interpretation of the hypothesis test

- **Two-Sided Test**: The two-sided test checks for **any difference** (either greater or less) between `FinalHealthScore` and `InitialHealthScore`. It answers the question, “Is there any significant difference between the two scores, regardless of direction?”
  
- **One-Sided Test**: The one-sided test focuses on a specific direction of the difference:
  - If we're testing whether the final score is **greater than** the initial score (as shown in the modified code), the p-value only reflects the probability that the final score exceeds the initial score. This tests a specific hypothesis: "Is the final score significantly **higher** than the initial score?"
  - If we test whether the final score is **less than** the initial score, the p-value would reflect only that direction.

### Whether or not we should indeed expect the p-value to be smaller in the "one tailed" versus "two tailed" analysis

Yes, in general, we should expect the p-value to be smaller in a **one-tailed test** than in a **two-tailed test** (assuming the data supports the directional hypothesis). This is because:

- In a **two-tailed test**, we are dividing the evidence between two extremes (greater than and less than), meaning the probability is split between both tails of the distribution.
- In a **one-tailed test**, we focus all the evidence on one direction (either greater than or less than). As a result, more of the probability mass is concentrated in one area, making the p-value smaller if the data strongly supports the direction you're testing.

### Summary:

- **Changes**: The p-value calculation now focuses on one side of the distribution (greater than or less than), depending on the direction of the hypothesis being tested.
- **Interpretation**: A one-sided test checks for differences in one direction (greater than or less than), whereas a two-sided test checks for any difference, regardless of direction.
- **p-value Size**: One-sided tests generally have smaller p-values than two-sided tests because they concentrate on a single direction.

- We discussed how these changes affect the interpretation of the hypothesis test:
  - A **two-sided test** checks for any difference in either direction.
  - A **one-sided test** checks only for a specific directional difference (greater or less than).

- We also talked about how, in general, a **one-sided test** produces a smaller p-value than a two-sided test if the data supports the directional hypothesis, as it focuses the probability mass on one side of the distribution.

## Fisher’s Tea Experiment Assignment: Statistical Analysis Report (Q8)

### **Problem Introduction**
This analysis evaluates the ability of STA130 students to correctly identify whether the milk or tea was poured first in a cup, similar to the famous tea experiment conducted by Ronald Fisher in the 1920s. In this experiment, 80 students from STA130 were asked to taste one cup of tea each and determine whether the milk or tea was poured first. Of these students, 49 were able to correctly identify the order. The goal is to determine if this result could be due to chance or if it provides evidence that students can reliably distinguish between the two pouring orders.

### **Relationship to Fisher’s Original Experiment**
Fisher's original experiment involved Dr. Muriel Bristol, who claimed she could distinguish the difference between tea poured first or milk poured first. Fisher devised an experiment to test whether her ability to identify the pouring order was due to chance or a genuine skill. In contrast, our experiment involves 80 students, which alters the scope and population. While Fisher tested one individual, our experiment generalizes the ability to an entire sample of students.

The key differences between the experiments are:
1. **Sample Size**: Fisher's original experiment involved only 8 cups, while this experiment includes 80 students tasting one cup each.
2. **Population**: Dr. Bristol’s ability might be considered more specialized, while the students represent a broader, less specialized population.

### **Statements of the Null and Alternative Hypotheses**
- **Null Hypothesis (H₀)**: The students are guessing randomly, and their ability to correctly identify whether the milk or tea was poured first is due to chance.
  - Formal version: \(H₀: p = 0.5 \), where \( p \) is the true proportion of students who can correctly identify the pouring order.
  - Informal version: There is no significant difference between students guessing correctly or incorrectly, and any observed difference from 50% is due to random guessing.
  
- **Alternative Hypothesis (H₁)**: More than 50% of the students can correctly identify whether the milk or tea was poured first, indicating they have some ability to distinguish between the two.
  - \( H₁: p > 0.5 \)

### **Quantitative Analysis**
To test the hypothesis, we will conduct a one-tailed z-test. Our observed data shows that 49 out of 80 students correctly identified the order.

- **Sample Size** \( n = 80 \)
- **Number of correct responses** \( x = 49 \)
- **Sample Proportion** \( p̂ = 49/80 = 0.6125 \)

The z-score is calculated as follows:

z = (p̂-$p_0$) / √($p_0$*(1-$p_0$)/n)

where \( $p_0$ = 0.5 \) is the null hypothesis proportion.

### **Methodology Code and Explanations**

````python
import numpy as np
from scipy import stats

# Set the seed for reproducibility
np.random.seed(42)

# Parameters
n = 80  # sample size
x = 49  # number of correct responses
p_hat = x / n  # observed sample proportion
p0 = 0.5  # null hypothesis proportion

# Standard error
se = np.sqrt(p0 * (1 - p0) / n)

# Z-score
z = (p_hat - p0) / se

# P-value (one-tailed test)
p_value = 1 - stats.norm.cdf(z)

# Output results
z, p_value
````

The output from the code is:
- **Z-Score**: 2.012
- **P-Value**: 0.0221

These results indicate that the observed proportion of students who correctly identified the pouring order is 2.012 standard deviations above the expected proportion under the null hypothesis. The p-value represents the probability of observing a result as extreme as this one under the assumption that the null hypothesis is true.

### **Findings and Discussion**
- **Z-Score Interpretation**: The z-score of 2.012 suggests that the observed proportion is somewhat unusual under the null hypothesis. It tells us that the proportion of students correctly identifying the pouring order is more than 2 standard deviations away from 0.5.
  
- **P-Value Interpretation**: The p-value of 0.0221 is below the common significance level \( α = 0.05 \), which means there is a 2.2% chance of observing such a result under the null hypothesis. This small probability provides evidence against the null hypothesis.

Since the p-value is less than 0.05, we reject the null hypothesis. This means that we have sufficient evidence to conclude that STA130 students, on average, have a better-than-random ability to distinguish whether the milk or tea was poured first.

### **Conclusion Regarding the Null Hypothesis**
Based on the analysis, we reject the null hypothesis. The proportion of students who correctly identified whether the milk or tea was poured first is significantly greater than 50%, suggesting that their ability to distinguish between the two is not due to random guessing. Thus, we can conclude that STA130 students have some ability to tell the difference between tea poured first and milk poured first.

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
