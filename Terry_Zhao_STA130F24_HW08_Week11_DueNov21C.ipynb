### Question 1

### (a) What is a Classification Decision Tree?

A **Classification Decision Tree** is a type of supervised learning algorithm that is used to address classification problems. These are problems where the goal is to predict the category or class of a given data point based on input features. 

It works by splitting the data into subsets based on feature values, aiming to create "branches" that lead to a prediction for a specific class. The splits are made to maximize the separation between different classes at each step, often using metrics like Gini Impurity or Information Gain.

#### Real-world Applications:
- **Medical Diagnosis**: Determining if a patient has a specific disease based on symptoms and test results.
- **Customer Segmentation**: Classifying customers into groups (e.g., high-value vs. low-value customers) for targeted marketing.
- **Fraud Detection**: Identifying whether a transaction is fraudulent based on patterns in the data.
- **Loan Approval**: Assessing whether a loan applicant is likely to repay based on their financial history.

---

### (b) Difference Between Predictions by Classification Decision Trees and Multiple Linear Regression (informal style):

Okay, so here's the deal. Classification Decision Trees and Multiple Linear Regression are like two different tools in your statistical toolbox, and they work on totally different kinds of problems.

- **Classification Decision Trees**: They’re all about answering "Which category does this thing belong to?" Imagine you're sorting your laundry—white clothes here, dark clothes there. The tree makes decisions at every step, splitting the data into smaller groups until it's confident enough to slap a label (or category) on it. It’s like, "Oh, this shirt is white because it’s made of cotton and doesn’t have any color patterns."

- **Multiple Linear Regression**: This is more like "How much of this thing will I get?" It's all about predicting a continuous number. Think of it as you trying to guess how much your Uber ride will cost based on distance and traffic. It’s just doing a fancy version of "plug these numbers into a formula and boom, here’s the answer."

Key difference: The tree works with categories (e.g., fraud/not fraud), while regression deals with actual numbers (e.g., $15 for the ride).

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

## Question 2

### **1. Accuracy**:

**Real-world Application**:  
**Weather forecasting for general conditions** (e.g., predicting "clear" vs. "rainy" days).

**Rationale**:  
Accuracy is great when all outcomes (true positives, true negatives, false positives, false negatives) are equally important. For weather, a wrong prediction of rain (when it's actually sunny) or vice versa has a similar level of inconvenience. So, focusing on overall correctness is suitable.

---

### **2. Sensitivity** (aka Recall):  

**Real-world Application**:  
**Disease screening tests** (e.g., detecting cancer in medical diagnostics).

**Rationale**:  
Sensitivity focuses on correctly identifying all actual positives. For diseases, it's critical to catch every potential case, even at the risk of false positives (FP). Missing a cancer diagnosis (FN) could have severe consequences, so a high sensitivity is crucial.

---

### **3. Specificity**:  

**Real-world Application**:  
**Spam email filters** (distinguishing spam vs. legitimate emails).

**Rationale**:  
Specificity emphasizes correctly identifying negatives. For spam filters, you want to avoid flagging legitimate emails as spam (FP). A high specificity ensures users don’t miss important non-spam emails.

---

### **4. Precision**:  

**Real-world Application**:  
**Fraud detection in banking** (identifying fraudulent transactions).

**Rationale**:  
Precision is about ensuring that predicted positives are truly positive. In fraud detection, falsely flagging a legitimate transaction as fraud (FP) can frustrate customers. A high precision ensures fewer false alarms while catching actual frauds.

#### **Step 1: Split the Dataset**
We split the data into 80% training and 20% testing using `train_test_split`. After splitting, the sizes of the training and test sets are:


````python
from sklearn.model_selection import train_test_split

# Split the dataset
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42
)

# Report the number of observations
train_size = len(ab_reduced_noNaN_train)
test_size = len(ab_reduced_noNaN_test)

print(f"Training set observations: {train_size}")
print(f"Testing set observations: {test_size}")
````

#### **Step 2: Fit the Decision Tree**
We train a decision tree model using the `List Price` variable to predict whether a book is "Hardcover" or "Paperback." Here’s the code to train the model and visualize it:

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Prepare the data
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])["H"]
X_train = ab_reduced_noNaN_train[["List Price"]]

# Train the DecisionTreeClassifier with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train)

# Visualize the decision tree
plt.figure(figsize=(10, 6))
tree.plot_tree(
    clf,
    feature_names=["List Price"],
    class_names=["Paperback", "Hardcover"],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Visualization")
plt.show()
````



#### **Step 3: Explain the Predictions**
Using the `tree.plot_tree()` visualization, we can interpret the decision tree:

1. **Root Node**:
   - The first decision splits books based on whether `List Price <= x` (e.g., `$25.50`).
   - If true, the prediction is a higher likelihood of "Paperback"; if false, it's "Hardcover."

2. **Intermediate Nodes**:
   - Further splits occur on different price thresholds (e.g., `$15.75`), refining the prediction for the "Paperback" category.

3. **Leaf Nodes**:
   - Each leaf provides the final prediction, indicating whether the book is classified as "Paperback" or "Hardcover" and showing the distribution of samples.

#### **Sample Decision Logic**:
- If `List Price <= $15.75`: Predict "Paperback."
- If `List Price > $15.75` and `<= $25.50`: Predict "Paperback."
- If `List Price > $25.50`: Predict "Hardcover."

The visualization clearly shows how the model uses thresholds in the `List Price` to predict the book type.

````python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score

# Confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Display confusion matrices
print("Confusion Matrix for clf:\n", cm_clf)
print("Confusion Matrix for clf2:\n", cm_clf2)
````

---

#### **Step 4: Define Metrics**
- **Accuracy**: (TP + TN)/(TP + TN + FP + FN)
- **Sensitivity** (Recall): TP/(TP + FN)
- **Specificity**:  TN/(TN + FP)

````python
# Function to calculate metrics
def calculate_metrics(cm):
    TP = cm[1, 1]
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]

    accuracy = (TP + TN) / (TP + TN + FP + FN)
    sensitivity = TP / (TP + FN)
    specificity = TN / (TN + FP)

    return np.round([accuracy, sensitivity, specificity], 3)

# Metrics for clf
metrics_clf = calculate_metrics(cm_clf)
# Metrics for clf2
metrics_clf2 = calculate_metrics(cm_clf2)

print("Metrics for clf: Accuracy, Sensitivity, Specificity =", metrics_clf)
print("Metrics for clf2: Accuracy, Sensitivity, Specificity =", metrics_clf2)
````

### Question 8

To visualize feature importances for a classification decision tree in scikit-learn, such as the model `clf2`, you can use the `.feature_importances_` attribute. This attribute provides the relative importance of each feature used in the decision tree. The importance values are based on how much each feature improves the model’s predictive accuracy through splits at each node, measured by criteria like Gini impurity or entropy.

Here's a step-by-step approach for visualizing feature importances:

1. **Access the feature importances**: After fitting the decision tree model (e.g., `clf2`), use `clf2.feature_importances_` to get the importance of each feature.
   
2. **Access feature names**: You can correlate these importances with feature names using `clf2.feature_names_in_`, which will give you a list of the feature names corresponding to the importance scores.

3. **Visualize the importances**: One way to do this is to create a bar plot, where the x-axis represents the feature names and the y-axis represents the corresponding importance score. You can use `matplotlib` or `seaborn` for this.

Here is an example in Python:

````python
import matplotlib.pyplot as plt
import numpy as np

# Assuming clf2 is your fitted decision tree model
feature_importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

# Create a bar plot of the feature importances
indices = np.argsort(feature_importances)[::-1]
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(len(feature_importances)), feature_importances[indices], align="center")
plt.xticks(range(len(feature_importances)), np.array(feature_names)[indices], rotation=90)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()

# Print the most important feature
most_important_feature = feature_names[indices[0]]
print(f"The most important feature for clf2 is: {most_important_feature}")
````

#### Interpretation

The most important feature for the clf2 model is "List Price," meaning that this predictor variable has the greatest influence on the decision-making process for classifying whether a book is a hardcover or paperback. This suggests that the model relies heavily on the price of the book to make predictions, potentially because books with higher prices are more likely to be hardcover, whereas lower-priced books are more likely to be paperback. The importance of "List Price" indicates that it is a strong distinguishing factor, which aligns with common industry practices where book pricing often correlates with binding type. This result can guide further analysis or business decisions, suggesting that focusing on price could be a key strategy in predicting book characteristics such as format.

In linear regression models, coefficients represent the direct relationship between each predictor and the outcome, where a larger coefficient indicates a stronger effect, assuming other variables are held constant.
