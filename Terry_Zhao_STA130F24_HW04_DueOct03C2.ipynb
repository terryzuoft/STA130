````python
import numpy as np

# Sample data (can replace with other datas)
sample = np.array([68, 75, 80, 85, 72, 90, 88, 95, 78, 82, 84, 77, 91, 89, 83, 70, 65, 87, 92, 76])

# Number of bootstrap resamples
n_bootstraps = 10000

# Function to calculate the bootstrap confidence interval for the mean
def bootstrap_ci_mean(sample, n_bootstraps=10000, ci_percentile=95):
    # Generate bootstrap samples and compute the mean for each
    bootstrapped_means = np.array([np.mean(np.random.choice(sample, size=len(sample), replace=True))
                                   for _ in range(n_bootstraps)])
    
    # Compute the lower and upper bounds of the confidence interval
    lower_bound = np.percentile(bootstrapped_means, (100 - ci_percentile) / 2)
    upper_bound = np.percentile(bootstrapped_means, 100 - (100 - ci_percentile) / 2)
    
    return lower_bound, upper_bound

# Calculate the 95% bootstrap confidence interval for the mean
ci_mean = bootstrap_ci_mean(sample)
print(f"95% Bootstrap Confidence Interval for the Mean: {ci_mean}")

# To calculate the 95% bootstrap confidence interval for a different statistic, 
# e.g., the median, simply replace np.mean with np.median in the function:

def bootstrap_ci_median(sample, n_bootstraps=10000, ci_percentile=95):
    # Generate bootstrap samples and compute the median for each
    bootstrapped_medians = np.array([np.median(np.random.choice(sample, size=len(sample), replace=True))
                                     for _ in range(n_bootstraps)])
    
    # Compute the lower and upper bounds of the confidence interval
    lower_bound = np.percentile(bootstrapped_medians, (100 - ci_percentile) / 2)
    upper_bound = np.percentile(bootstrapped_medians, 100 - (100 - ci_percentile) / 2)
    
    return lower_bound, upper_bound

# Calculate the 95% bootstrap confidence interval for the median
ci_median = bootstrap_ci_median(sample)
print(f"95% Bootstrap Confidence Interval for the Median: {ci_median}")
````

## Problem Introduction 

In this context of AliTech, we are comparing the initial health score with the final health score after taking the vaccine.

**Null Hypothesis**: The vaccine does not work. It cannot affect one's health score. 

**Alternative Hypothesis**: The vaccine does work. It has a positive effect on one's health score. 

### Data Visualization (Obtain From Code Below)



## Quantitative Analysis

### Methodology Code with Comments + Supporting Visualization

Note: I have created a .csv file on notepad and uploaded it to my GitHub. Also, I have also took screenshots of the result the code have produced and included it in the markdown as fig.show(renderer="png") often does not function properly 

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Reproducibility
np.random.seed(42)

# Reading the CSV data directly from GitHub raw URL
url = 'https://raw.githubusercontent.com/terryzuoft/STA130/main/vaccine_data.csv'
data = pd.read_csv(url)

# Calculate the difference in health scores
data['ScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Data Visualization
plt.figure(figsize=(10, 5))
sns.boxplot(data=data[['InitialHealthScore', 'FinalHealthScore']])
plt.title('Initial vs Final Health Scores')
plt.show()

# Bootstrapping function
def bootstrap(data, n_bootstrap=1000):
    bootstrap_means = []
    for i in range(n_bootstrap):
        sample = data.sample(frac=1, replace=True)
        mean_diff = sample['FinalHealthScore'].mean() - sample['InitialHealthScore'].mean()
        bootstrap_means.append(mean_diff)
    return bootstrap_means

# Run bootstrapping
bootstrap_means = bootstrap(data)

# Plot bootstrapping results
plt.hist(bootstrap_means, bins=30, edgecolor='black')
plt.axvline(np.mean(bootstrap_means), color='red', linestyle='dashed', linewidth=2)
plt.title('Bootstrapping Mean Differences')
plt.show()

# Findings: Calculate p-value and confidence intervals
observed_mean_diff = data['ScoreDifference'].mean()
p_value = np.mean(np.array(bootstrap_means) >= observed_mean_diff)
print(f"Observed Mean Difference: {observed_mean_diff}")
print(f"p-value: {p_value}")
````

## Findings and Dicussion + Quantitative Analysis
