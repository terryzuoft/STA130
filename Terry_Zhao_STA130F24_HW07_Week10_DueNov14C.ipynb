### Question 1

**1. Difference between Simple Linear Regression and Multiple Linear Regression:**
   - **Simple Linear Regression** uses a single predictor (independent) variable to model the relationship with a single outcome (dependent) variable. The model takes the form \( y = \beta_0 + \beta_1 x + \epsilon \), where \( x \) is the predictor.
   - **Multiple Linear Regression** uses two or more predictors to model the outcome variable. This model takes the form \( y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon \).
   - **Benefit of Multiple Linear Regression:** It allows for a more comprehensive understanding of the outcome by accounting for additional variables, potentially leading to a more accurate model of complex relationships.

---

**2. Difference between Continuous and Indicator Variables in Simple Linear Regression:**
   - A **continuous variable** can take on any numerical value within a range (e.g., age, height). When used in linear regression, it models a continuous linear relationship.
   - An **indicator variable** (also called a dummy variable) is binary, taking values such as 0 and 1, indicating categories or groups (e.g., gender: 0 for male, 1 for female). It models discrete jumps rather than a continuous trend.

---

**3. Change in Model Behavior with an Indicator and Continuous Variable in Multiple Linear Regression:**
   - When both a **continuous** and an **indicator variable** are included in a Multiple Linear Regression model, the indicator variable introduces a shift in the intercept for each category it represents. The model can now account for differences in the outcome based on the categories of the indicator while adjusting for the continuous variable.
   - For example, the model \( y = \beta_0 + \beta_1 x + \beta_2 d + \epsilon \) where \( x \) is continuous and \( d \) is the indicator, captures different baseline levels for each category defined by \( d \), while allowing \( x \) to affect the outcome across all categories.

---

**4. Effect of Adding an Interaction between Continuous and Indicator Variables in Multiple Linear Regression:**
   - Adding an **interaction term** (e.g., \( y = \beta_0 + \beta_1 x + \beta_2 d + \beta_3 x \cdot d + \epsilon \)) allows the effect of the continuous variable on the outcome to differ between categories.
   - This interaction changes the slope of the continuous predictor for different levels of the indicator variable, modeling a different rate of change in \( y \) for each group defined by the indicator.

---

**5. Behavior of a Multiple Linear Regression Model Based on Indicator Variables from a Non-Binary Categorical Variable:**
   - When a **non-binary categorical variable** (e.g., three levels: low, medium, high) is used, it is encoded with **multiple indicator (dummy) variables** to represent each category (e.g., two dummy variables to represent three categories).
   - This model takes the form \( y = \beta_0 + \beta_1 d_1 + \beta_2 d_2 + \epsilon \), where each \( d \) represents a level of the categorical variable. These dummy variables are binary (0 or 1) for each category, enabling the model to capture different baselines for each level of the categorical variable without assuming a continuous relationship.

**1. Identify the Outcome and Predictor Variables:**
   - The **outcome (dependent) variable** is likely **sales** or **revenue** generated from selling sports equipment, as this is the primary business measure the company wants to improve through advertising.
   - The **predictor (independent) variables** are the **advertising budgets** for:
     - **TV ads**
     - **Online ads**
   - Both advertising budgets are continuous in this scenario, meaning they represent the dollar amount spent on each type of ad.

---

**2. Considering Interaction Effects:**
   - There may be an **interaction effect** between the TV and online ad budgets, where the impact of TV advertising on sales could be influenced by the amount spent on online advertising and vice versa. This is because customers exposed to ads through both mediums might have a higher likelihood of purchasing than those exposed through only one.
   - To capture this relationship, an **interaction term** between the TV and online advertising budgets should be considered in the model.

---

**3. Linear Forms with and without Interaction:**

   - **Without Interaction (Simple Multiple Linear Regression):**
     - The model without interaction simply considers the separate effects of each advertising budget:
       \[
       \text{Sales} = \beta_0 + \beta_1 (\text{TV Budget}) + \beta_2 (\text{Online Budget}) + \epsilon
       \]
     - Here, \( \beta_1 \) represents the effect of the TV advertising budget on sales, assuming no relationship with the online budget, and \( \beta_2 \) represents the effect of the online advertising budget on sales, assuming no relationship with the TV budget.

   - **With Interaction (Multiple Linear Regression with Interaction Term):**
     - Including an interaction term allows us to account for the interdependent effects of the advertising budgets:
       \[
       \text{Sales} = \beta_0 + \beta_1 (\text{TV Budget}) + \beta_2 (\text{Online Budget}) + \beta_3 (\text{TV Budget} \times \text{Online Budget}) + \epsilon
       \]
     - Here, \( \beta_3 \) captures the interaction effect, adjusting the relationship between each budget and sales based on the level of the other budget.

---

**4. Using the Models to Make Predictions and Difference Between Models:**
   - **Without Interaction:** When predicting sales without the interaction term, each type of advertising is assumed to have an independent effect. For instance, a dollar spent on TV advertising increases sales by the same amount regardless of the online ad budget, and vice versa.
   - **With Interaction:** In this model, the effect of the TV budget on sales depends on the online budget, and vice versa. For example, if both budgets are high, the combined effect might be larger than if they were spent separately. This model could provide more accurate predictions if the mediums indeed influence each other.

---

**5. Modifying the Models for Binary Predictor Variables:**

   - **With Binary Predictors (No Interaction):**
     - Suppose each budget is categorized as **high (1)** or **low (0)**. The model without interaction would look like this:
       \[
       \text{Sales} = \beta_0 + \beta_1 (\text{TV High/Low}) + \beta_2 (\text{Online High/Low}) + \epsilon
       \]
     - Here, \( \beta_1 \) represents the average difference in sales between low and high TV budgets, assuming no interaction with the online budget, and \( \beta_2 \) does the same for online ads.

   - **With Binary Predictors and Interaction:**
     - Adding an interaction term with binary variables would yield:
       \[
       \text{Sales} = \beta_0 + \beta_1 (\text{TV High/Low}) + \beta_2 (\text{Online High/Low}) + \beta_3 (\text{TV High/Low} \times \text{Online High/Low}) + \epsilon
       \]
     - In this model, \( \beta_3 \) represents the extra effect on sales when both budgets are high compared to the sum of their individual effects.

---

**Summary:**
The model without interaction assumes each advertising type has an independent effect on sales, while the model with interaction captures a potential amplifying effect when both advertising budgets are high. Using binary predictor variables instead of continuous ones simplifies the model, but may reduce precision by not capturing the nuanced effects of varying budget amounts.

```python
     import statsmodels.formula.api as smf
```

     ```python
     logit_interaction = smf.logit("Outcome ~ Continuous_Var * Binary_Var", data=data).fit()
     ```

     - The intercept \(\beta_0\) is the baseline level of the outcome.
     - Each coefficient (\(\beta_1, \beta_2\), etc.) represents the expected increase in the outcome per unit increase in the predictor variable.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import statsmodels.formula.api as smf

# Define lists to store results
in_sample_mse = []
out_of_sample_mse = []

# Run the loop for a set number of iterations
for i in range(100):  # Adjust the range for more or fewer iterations
    # Split data randomly in each iteration (no fixed random seed here)
    train_data, test_data = train_test_split(pokeaman_train, test_size=0.2)
    
    # Define and fit the model on the training data
    model = smf.ols(formula=model4_linear_form_CS, data=train_data).fit()
    
    # Calculate in-sample (training) MSE
    in_sample_pred = model.predict(train_data)
    in_sample_mse.append(mean_squared_error(train_data['HP'], in_sample_pred))
    
    # Calculate out-of-sample (test) MSE
    out_of_sample_pred = model.predict(test_data)
    out_of_sample_mse.append(mean_squared_error(test_data['HP'], out_of_sample_pred))

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(in_sample_mse, label='In-Sample MSE', color='blue')
plt.plot(out_of_sample_mse, label='Out-of-Sample MSE', color='red')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.title('In-Sample vs. Out-of-Sample MSE across Multiple Iterations')
plt.show()
````
